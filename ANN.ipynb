```python
#packages
from sklearn import datasets
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#Telecharger la base de données
df=datasets.load_breast_cancer()
```

```python
#decouvrir la base de données
print(df.DESCR)
```

```python
#Notre target
df.target
```

```python
df_pd = pd.DataFrame(df.data, columns=df.feature_names)
df_pd.head()
```

```python
#preparer les donnee
X=df.data   #notre data
y=df.target   #target 0 ou 1
```

```python
scaler =StandardScaler()
X_scaled=scaler.fit_transform(X)   #standarisé notre donné et transformé
X_scaled
```

```python
#divise notre donner a des donne de train (80%) et de donner de test
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
#constructionde la reseau de Neurone
model=tf.keras.models.Sequential()
```

```python
#ajouter une layout hidden a notre model
model.add(tf.keras.layers.Dense(units=15,activation='relu',kernel_initializer='uniform',input_dim=30))
```

```python
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid',kernel_initializer='uniform'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['binary_accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
#teste notre model
test_loss,test_accarary=model.evaluate(X_test,y_test)
print(f'Test accuray {test_accarary} et test loss {test_loss}')
```
